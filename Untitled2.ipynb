```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import zipfile
import os

# To display all columns without truncation
pd.set_option('display.max_columns', None)

# Step 2: Load the dataset

# Check if the zip file exists and extract it
zip_file_path = '/content/archive (1).zip'
extracted_dir = '/content/'

if os.path.exists(zip_file_path):
    print(f"Extracting {zip_file_path}...")
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extracted_dir)
    print(f"Contents extracted to {extracted_dir}")

    # List files to see what was extracted
    print("Files in extracted directory:")
    for file_name in os.listdir(extracted_dir):
        print(file_name)

# Assuming the CSV file is named 'Unemployment in India.csv' inside the zip or after extraction
data = pd.read_csv(os.path.join(extracted_dir, "Unemployment in India.csv"))

# Display first 5 rows
print("First 5 Rows:\n")
print(data.head())

# Check info
print("\nDataset Information:\n")
print(data.info())

# Check for missing values
print("\nMissing Values:\n")
print(data.isnull().sum())

# Step 3: Data Cleaning and Preprocessing

# 3A. Clean column names (remove spaces)
data.columns = data.columns.str.strip()

# 3B. Drop missing values (since all missing rows are the same)
data.dropna(inplace=True)

# 3C. Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'].str.strip(), format='%d-%m-%Y')

# 3D. Verify the cleaning
print("âœ… After Cleaning â€” Dataset Info:\n")
print(data.info())

print("\nâœ… Missing Values After Cleaning:\n")
print(data.isnull().sum())

# Display first 5 cleaned rows
print("\nâœ… Cleaned Data (First 5 Rows):\n")
print(data.head())

# Step 4A: Basic statistical overview
print("ðŸ“Š Basic Statistical Summary:\n")
print(data.describe())

# Step 4B: Plot average unemployment trend over time
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(data['Date'], data['Estimated Unemployment Rate (%)'], color='orange')
plt.title('ðŸ“ˆ National Unemployment Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.grid(True)
plt.show()

# Step 4C: Average unemployment rate by region
region_avg = data.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
region_avg.plot(kind='bar', color='skyblue')
plt.title('ðŸ™ï¸ Top 10 Regions by Average Unemployment Rate')
plt.ylabel('Average Unemployment Rate (%)')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.show()

# Step 4D: Rural vs Urban comparison
area_avg = data.groupby('Area')['Estimated Unemployment Rate (%)'].mean()

plt.figure(figsize=(5,5))
area_avg.plot(kind='bar', color=['lightgreen', 'salmon'])
plt.title('ðŸ¡ Rural vs Urban Unemployment Rate')
plt.ylabel('Average Unemployment Rate (%)')
plt.xlabel('Area Type')
plt.show()

print("\nAverage Unemployment by Area:\n", area_avg)

import matplotlib.pyplot as plt
import seaborn as sns

# Set style for charts
sns.set(style="whitegrid")

# Calculate average unemployment rate by region
region_unemployment = data.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)

# Plot the average unemployment rate by region
plt.figure(figsize=(12,6))
sns.barplot(x=region_unemployment.values, y=region_unemployment.index, palette='coolwarm')
plt.title("Average Unemployment Rate by Region", fontsize=16)
plt.xlabel("Average Unemployment Rate (%)")
plt.ylabel("Region")
plt.show()

# Group by date and calculate mean unemployment rate across all regions
date_trend = data.groupby('Date')['Estimated Unemployment Rate (%)'].mean()

# Plot the unemployment rate trend over time
plt.figure(figsize=(12,6))
plt.plot(date_trend.index, date_trend.values, color='blue', linewidth=2)
plt.title("Unemployment Rate Trend Over Time (India Average)", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()

# Average unemployment by area type (Rural/Urban)
area_comparison = data.groupby('Area')['Estimated Unemployment Rate (%)'].mean()

# Bar chart
plt.figure(figsize=(6,5))
sns.barplot(x=area_comparison.index, y=area_comparison.values, palette='viridis')
plt.title("Average Unemployment Rate: Rural vs Urban", fontsize=16)
plt.xlabel("Area")
plt.ylabel("Average Unemployment Rate (%)")
plt.show()

top5 = region_unemployment.head(5)
bottom5 = region_unemployment.tail(5)

print("ðŸ† Top 5 Regions with Highest Unemployment:\n")
print(top5)

print("\nðŸŒ¿ Bottom 5 Regions with Lowest Unemployment:\n")
print(bottom5)


# Compute correlation matrix for numeric columns
corr = data[['Estimated Unemployment Rate (%)',
           'Estimated Employed',
           'Estimated Labour Participation Rate (%)']].corr()

# Plot the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap", fontsize=16)
plt.show()

sns.pairplot(data[['Estimated Unemployment Rate (%)',
                 'Estimated Employed',
                 'Estimated Labour Participation Rate (%)']],
             diag_kind='kde', palette='husl')

plt.suptitle("Pairwise Relationships Between Key Variables", y=1.02)
plt.show()

plt.figure(figsize=(12,6))
sns.scatterplot(data=data,
                x='Estimated Employed',
                y='Estimated Unemployment Rate (%)',
                hue='Area', style='Area', alpha=0.7)
plt.title("Employment vs Unemployment Rate by Area Type", fontsize=16)
plt.xlabel("Estimated Employed")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.legend(title='Area')
plt.show()

plt.figure(figsize=(14,7))
for region in data['Region'].unique():
    region_data = data[data['Region'] == region]
    plt.plot(region_data['Date'], region_data['Estimated Unemployment Rate (%)'], alpha=0.5)

plt.title("Unemployment Rate Trend per Region (2019â€“2020)", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```
